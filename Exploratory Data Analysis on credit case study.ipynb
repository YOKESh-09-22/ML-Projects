```python
appli_data_target0 = df[df['TARGET']==0]
appli_data_target0.head()
```

```python
appli_data_target1 = df[df['TARGET']==1]
appli_data_target1.head()
```

```python
cols_num=list(appli_data_target0.select_dtypes('int64').columns)
cols_float=list(appli_data_target0.select_dtypes('float').columns)

cols=cols_num+cols_float

Nondef_num=appli_data_target0[cols]
Nondef_corr = Nondef_num.corr()
round(Nondef_corr, 3)
```

```python
l1=Nondef_corr.unstack()
l1.sort_values(ascending=False).drop_duplicates()
```

```python
cols_num=list(appli_data_target1.select_dtypes('int64').columns)
cols_float=list(appli_data_target1.select_dtypes('float').columns)

cols=cols_num+cols_float

def_num=appli_data_target1[cols]
def_corr = def_num.corr()
round(def_corr, 3)
```

```python
l1=def_corr.unstack()
l1.sort_values(ascending=False).drop_duplicates()
```

# 10. Final Conclusion

1.GOODS_PRICE loans are provided on the basis of AMT_ANNUITY.

2.With low credit persons more loans have been provided.

3.More people having difficulties in repaying always prefer to buy Cash loans.

4.Non-car owners always prefer to buy revolving loans.

5.Male Target returning count is higher compared to females.

6.We observe a decrease in the percentage of married and widowed with Loan Payment Difficulties and an increase in the percentage of single and civil married with Loan Payment Difficulties when compared with the percentages of both Loan Payment Difficulties and Loan Non-Payment Difficulties.
